#Cleaning data

```python
#removing data with 0 solar radiation before 6am and after 6pm
clean1=df[df['Hour']<6]
clean2=df[df['Hour']>18]
```

```python
df=df.drop((clean1.index | clean2.index),axis=0)
```

```python
df.info()
```

```python
x=df.drop('Solar Radiation (GHI)',axis=1)
y=df['Solar Radiation (GHI)']
```

```python
corr_matrix=df.corr()
```

```python
features=corr_matrix.index
features
```

```python
plt.figure(figsize=(10,10))
heatmap=sns.heatmap(df[features].corr(),annot=True)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
from sklearn.preprocessing import MinMaxScaler
x=pd.DataFrame(MinMaxScaler().fit(x).transform(x))
x.head()
```

```python
best_feature=SelectKBest(score_func=chi2,k=10)
fit=best_feature.fit(x,y)
```

```python
scores=pd.DataFrame(fit.scores_)
col=pd.DataFrame(features)
```

```python
feature_score=pd.concat([scores,col],axis=1)
feature_score.columns=['Score','Feature']
feature_score.head()
```

```python
feature_score.nlargest(10,'Score')
```

```python
feature_score.nlargest(10,'Score').plot(kind='bar')
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model.fit(x,y)
```
